```python
import re
opinions_text = s.splitlines()
pattern = r"^(GOOD|BAD) (.*)"
opinions = []

for ol in opinions_text:
    m = re.match(pattern, ol)
    if m:
        group_1 = m.group(1)
        group_2 = m.group(2)
        opinions.append((group_1 == "GOOD", group_2))
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, set_seed
from torch.nn import functional as F
import random

model_name = 'flax-community/papuGaPT2'
device = 'cuda'
device = 'cpu'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)

def log_probs_from_logits(logits, labels):
    logp = F.log_softmax(logits, dim=-1)
    logp_label = torch.gather(logp, 2, labels.unsqueeze(2)).squeeze(-1)
    return logp_label


def sentence_prob(sentence_txt):
    input_ids = tokenizer(sentence_txt, return_tensors='pt')['input_ids'].to(device)
    with torch.no_grad():
        output = model(input_ids=input_ids)
        log_probs = log_probs_from_logits(output.logits[:, :-1, :], input_ids[:, 1:])
        seq_log_probs = torch.sum(log_probs)
    return seq_log_probs.cpu().numpy()
```

```python
import random

generator = pipeline('text-generation', model='flax-community/papuGaPT2')

print('Model loaded')

MAX_TOKENS_NUM = 100
correct_answers = 0
total = 0

random_opinions = random.sample(opinions, 50)
print(random_opinions)
print(len(random_opinions))
for label, opinion in random_opinions:
    prompt_positive = opinion + " to zdanie pozytywne"
    prompt_negative = opinion + " to zdanie negatywne"
    if len(prompt_positive) > MAX_TOKENS_NUM:
        continue
    g_positive = generator(prompt_positive, pad_token_id=generator.tokenizer.eos_token_id, max_new_tokens = MAX_TOKENS_NUM)[0]['generated_text']
    g_negative = generator(prompt_negative, pad_token_id=generator.tokenizer.eos_token_id, max_new_tokens = MAX_TOKENS_NUM)[0]['generated_text']
    score_positive = sentence_prob(g_positive)
    score_negative = sentence_prob(g_negative)
    answer = score_positive > score_negative
    if answer == label:
        correct_answers += 1
    total += 1
    print(f"Current accuracy: {correct_answers / total}")



```
